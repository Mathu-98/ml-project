# **Replace Missing Values and Remove Duplicates**

```python
df['Customer_Name'].fillna('Not Provided', inplace=True)
```

```python
cols_to_fill = ['Email', 'City', 'Country', 'Region', 'Product_Name', 'Category', 'Status', 'Delivery_Mode', 'Driver_Name', 'Warehouse_Name', 'Manager_Name', 'Order_Priority', 'Payment_Mode', 'Insurance', 'Return_Status','Delivery_Center' ]
# Fill missing values with "Unknown"
df[cols_to_fill] = df[cols_to_fill].fillna('Unknown')
```

```python
df['Weight_kg'] = df['Weight_kg'].fillna(df.groupby('Category')['Weight_kg'].transform('mean'))
df['Fuel_Cost'] = df['Fuel_Cost'].fillna(df.groupby('Transport_Type')['Fuel_Cost'].transform('median'))
df['Remarks'] = df['Remarks'].fillna("No Remarks")
```

```python
mode_value = df['Transport_Type'].mode()[0]
df['Transport_Type'].fillna(mode_value, inplace=True)
```

```python
df.loc[df['Quantity'].isna() & df['Unit_Price'].notna() & df['Total_Price'].notna(), 'Quantity'] = df['Total_Price'] / df['Unit_Price']
df.loc[df['Unit_Price'].isna() & df['Quantity'].notna() & df['Total_Price'].notna(), 'Unit_Price'] = df['Total_Price'] / df['Quantity']
df.loc[df['Total_Price'].isna() & df['Quantity'].notna() & df['Unit_Price'].notna(), 'Total_Price'] = df['Quantity'] * df['Unit_Price']
```

```python
df['Quantity'].fillna(df['Quantity'].median(), inplace=True)
df['Unit_Price'].fillna(df['Unit_Price'].median(), inplace=True)
df['Total_Price'].fillna(df['Total_Price'].median(), inplace=True)
```

```python
df['Fuel_Cost'].fillna(df['Fuel_Cost'].mean(), inplace=True)
df['Distance_km'].interpolate(method='linear', inplace=True)
```

```python
df.isnull().sum()
```

```python
# Drop duplicates based on Shipment_ID and Order_ID
df.drop_duplicates(subset=['Shipment_ID', 'Order_ID'], inplace=True)
```

## **Drop Invalid values and Outliers**

```python
df = df.drop(df[(df['Quantity'] <= 0) |
    (df['Weight_kg'] <= 0) |
    (df['Total_Price'] <= 0) |
    (df['Delivery_Date'] <= df['Shipment_Date']) |
    (df['Delay_Days'] < -1)].index)
```

```python
def drop_outliers(df, columns):
    """
    Drops outliers from specified columns using IQR method.
    """
    for col in columns:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower = Q1 - 1.5 * IQR
        upper = Q3 + 1.5 * IQR
        df = df[(df[col] >= lower) & (df[col] <= upper)]
    return df

# Example usage
cols_to_clean = ['Weight_kg', 'Quantity', 'Unit_Price', 'Total_Price']
cleaned_df = drop_outliers(df, cols_to_clean)
```

```python
cleaned_df["Payment_Mode"] = cleaned_df["Payment_Mode"].replace({"Credt":"Credit","Debet":"Debit"})
cleaned_df["Status"] = cleaned_df["Status"].replace({ "delayed": "Delayed", "DELAYED": "Delayed" })
cleaned_df["Country"] = cleaned_df["Country"].replace({ "us": "USA", "U.K.": "UK" })
df = cleaned_df.copy()
```

```python
df = df.drop(df[df['Distance_km'] > 20000].index)
```

```python
Q1 = df['Distance_km'].quantile(0.25)
Q3 = df['Distance_km'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
df_no_outliers = df[(df['Distance_km'] >= lower_bound) &
 (df['Distance_km'] <= upper_bound)]
print(df_no_outliers)
```

```python
import numpy as np
email_pattern = r'^[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Za-z]{2,}$'
# Replace invalid email formats with NaN
df['Email'] = df['Email'].where(df['Email'].str.match(email_pattern, na=False), np.nan)
print(df)
```

```python
import plotly.express as px
import plotly.graph_objects as go
def logistics_dashboard_interactive(df):
  # 1. Shipment Volume by Region
  fig1 = px.bar(df['Region'].value_counts().reset_index(),
              x='Region', y='count',
              title="Shipment Volume by Region",
              labels={'Region':'Region','count':'Count'})
  fig1.show()
  # 2. Delivery Status Distribution
  fig2 = px.pie(df, names='Status', title="Delivery Status Distribution")
  fig2.show()
  # 3. Delay Days by Region
  fig3 = px.box(df, x='Region', y='Delay_Days', title="Delay Days by Region")
  fig3.show()
  # 4. Profit vs Distance
  fig4 = px.scatter(df, x='Distance_km', y='Profit', color='Order_Priority',
                  hover_data=['Shipment_ID','Customer_Name'],
                  title="Profit vs Distance (colored by Order Priority)")
  fig4.show()
  # 5. Fuel Efficiency by Transport Type
  fig5 = px.bar(df, x='Transport_Type', y='Fuel_Efficiency', title="Fuel Efficiency by Transport Type")
  fig5.show()
  # 6. Revenue Contribution by Category
  category_revenue = df.groupby('Category')['Revenue'].sum().reset_index()
  fig6 = px.bar(category_revenue,
              x='Category', y='Revenue',
              title="Revenue Contribution by Category")
  fig6.show()
# Run the interactive dashboard
logistics_dashboard_interactive(df)
```
